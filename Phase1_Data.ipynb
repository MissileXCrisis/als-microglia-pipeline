## Looking at the raw data 

```python
import h5py
import numpy as np

# Pick the first raw file from your unzipped list
h5_file_path = "data/raw/GSM6186103_MA001_filtered_feature_bc_matrix.h5"

print(f"--- INSPECTING HDF5 STRUCTURE FOR: {h5_file_path} --- \n")

# Open the binary file in read-only mode
with h5py.File(h5_file_path, 'r') as f:
    # 1. Walk through the internal group structure recursively
    def print_structure(name, obj):
        if isinstance(obj, h5py.Dataset):
            print(f"  Dataset: {name} | Shape: {obj.shape} | Type: {obj.dtype}")
        elif isinstance(obj, h5py.Group):
            print(f"  Group: {name}")

    f.visititems(print_structure)
    
    print("\n--- PEEPING ACTUAL RAW DATA SNIPPETS ---")
    
    # 2. Extract a few real barcodes (Cell IDs)
    barcodes = f['matrix/barcodes'][:5]
    print(f"\nFirst 5 Cell Barcodes:\n{[b.decode('utf-8') for b in barcodes]}")
    
    # 3. Extract a few real Gene Names
    gene_names = f['matrix/features/name'][:5]
    print(f"\nFirst 5 Gene Symbols:\n{[g.decode('utf-8') for g in gene_names]}")
    
    # 4. Check matrix dimensions
    matrix_shape = f['matrix/shape'][:]
    print(f"\nMatrix Shape Array [Genes, Cells]: {matrix_shape}")
```

# Converting Raw Metadata into CSV 

```python
import gzip
import pandas as pd

matrix_file = "data/metadata/GSE204704_series_matrix.txt.gz"
output_file = "data/metadata/raw_horizontal_metadata.csv"

raw_rows = []

# 1. Read the compressed file and extract ONLY sample metadata lines
with gzip.open(matrix_file, 'rt') as f:
    for line in f:
        if line.startswith("!Sample_"):
            # Split by tabs and strip quotes to keep the raw text clean
            split_line = [item.strip('"') for item in line.strip().split('\t')]
            raw_rows.append(split_line)

# 2. Load into a DataFrame without changing the horizontal shape
# The first column will be the GEO Row Header (e.g., !Sample_title)
df_raw_grid = pd.DataFrame(raw_rows)

# 3. Save it as a standard flat CSV file
df_raw_grid.to_csv(output_file, index=False, header=False)

print(f"🎉 Extracted! Raw metadata saved to: {output_file}")
print(f"Shape of the raw grid: {df_raw_grid.shape[0]} rows by {df_raw_grid.shape[1]} columns.")
```

# Parse and Ingest into SQL

```python
import os
import gzip
import sqlite3
import pandas as pd

matrix_file = "data/metadata/GSE204704_series_matrix.txt.gz"

# Initialize explicit data lists
gsm_ids = []
titles = []
tissues = []
antibodies = []
h5_filenames = []

with gzip.open(matrix_file, 'rt') as f:
    for line in f:
        # 1. Pull core identifiers
        if line.startswith("!Sample_title"):
            titles = [x.strip('"') for x in line.split('\t')[1:]]
        elif line.startswith("!Sample_geo_accession"):
            gsm_ids = [x.strip('"') for x in line.split('\t')[1:]]
            
        # 2. Target specific characteristics rows dynamically
        elif line.startswith("!Sample_characteristics_ch1"):
            elements = [x.strip('"') for x in line.split('\t')[1:]]
            # Check if this line holds tissue data
            if ":" in elements[0] and "tissue" in elements[0]:
                tissues = [x.split(":")[-1].strip() for x in elements]
            # Check if this line holds antibody/cell-hashing data
            elif any("antibodies" in x for x in elements):
                antibodies = [x.split(":")[-1].strip() if "antibodies" in x else "Single-Run" for x in elements]
                
        # 3. Pull the actual .h5 filenames out of the download URLs
        elif line.startswith("!Sample_supplementary_file_1"):
            urls = [x.strip('"') for x in line.split('\t')[1:]]
            h5_filenames = [os.path.basename(url) for url in urls]

# Fallback check in case antibody row was empty for some columns
if not antibodies:
    antibodies = ["Single-Run"] * len(gsm_ids)

# Build unified dataframe
df_clean = pd.DataFrame({
    'gsm_id': gsm_ids,
    'sample_title': titles,
    'tissue': tissues,
    'batch_protocol': antibodies,
    'h5_filename': h5_filenames
})

# 4. Feature Engineering: Extract Donor ID from title (e.g. "ALS4 GM MA001" -> "ALS4")
df_clean['donor_id'] = df_clean['sample_title'].apply(lambda x: x.split()[0] if len(x.split()) > 0 else "Unknown")

print("--- PIPELINE VERIFICATION METRICS ---")
print(f"Total Samples Indexed: {len(df_clean)}")
print(f"Columns Created: {list(df_clean.columns)}")

# Save to local SQLite
conn = sqlite3.connect("data/als_microglia.db")
df_clean.to_sql("cohort_metadata", conn, if_exists="replace", index=False)
conn.close()
print("\n🎉 Phase 1 Complete! Database is fully optimized for downstream data processing.")
```
